### Q3: Most common programming language

Print the value of the single most common programming language (`programming_languages` column).

Example:
```
"Most common language: Python"
```

````python
# Your code here
import collections
lang = ['R','Wolfram Mathematica','Matlab/Octave','JavaScript','Python','Julia','Java','Scala','C/C++',]
langs = data.programming_languages
l1 = [n.split(',') for n in langs]
xx=l1[1]+l1[0]
y=[]
for i in range(len(l1)):
    y = y+l1[i]
yy=[]
for i in range(len(y)):
    yy.append(y[i].strip())
df= pd.Series(yy)
freq = df.value_counts().idxmax()
print('Most common language: ', freq)
````

### Q4: Average Python Proficiency

Calculate the average python proficiency (`python_proficiency`) given previous experience working with neural networks (`neural_networks_experience`).

Show the results in a bar plot.

Example:



````python
# Your code here
neural = data.neural_networks_experience
py = data.python_proficiency
n_sum = 0 
y_sum = 0
c_n = 0
c_y = 0
for i in range(len(py)): 
    if neural.iloc[i] == 'No':
        c_n=c_n+1
        n_sum = n_sum + py[i]
        n= n_sum/c_n
    if neural.iloc[i] == 'Yes':
        c_y=c_y+1
        y_sum = y_sum + py[i] 
        y = y_sum/c_y

avg_x= ['Yes', 'No']
avg_y = [n,y] # average of yes and no

#Plot

plt.figure(figsize=(10,6))
plt.bar(avg_x, avg_y) 
plt.title('Average python proficiency given neural network experience')
plt.xlabel('neural_networks_experience')
plt.ylabel('python_proficiency')
plt.show()
````

### Q5: Working students not interested in academia

Calculate what's the percetage of students without academic interest (`academic_interest`) that are currently working (`currently_working`).

Example:

```
 Answer: 50% of the students that are not interested in academics are currently working.
```

````python
# Your code here
aca = data.academic_interest
work = data.currently_working

a = [n for n in range(len(aca)) if aca.iloc[n]=='No' or aca.iloc[n]=='Maybe']
w = [n for n in range(len(work)) if work.iloc[n] == 'Yes']
# a,w
res=0
for i in range(len(a)):
    for j in range(len(w)):
        if a[i] == w[j]:
            res=res+1
    else:
        pass
res
print('Answer:',res/len(aca)*100,'% of the students that are not interested in academics are currently working.')
````
